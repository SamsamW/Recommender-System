```python
import os
import json

# 假设你所有的 JSON 文件都在同一个目录中
directory_path = '/Users/jianing/Documents/Study/Recsys/Hands-On-Recommender-System/recommendations/gpt_data/'

# 准备一个空的列表来存储合并后的数据
merged_data = []

# 遍历目录中的所有 JSON 文件
for filename in os.listdir(directory_path):
    if filename.endswith('.json'):
        file_path = os.path.join(directory_path, filename)
        
        # 获取城市名称（假设文件名格式为 City_travel_data.json）
        city_name = filename.split('_')[0]
        
        # 读取 JSON 文件内容
        with open(file_path, 'r') as json_file:
            data = json.load(json_file)
        
        # 添加 city 特征
        if isinstance(data, list):
            for item in data:
                item['city'] = city_name
            # 将数据追加到合并后的列表中
            merged_data.extend(data)
        elif isinstance(data, dict):
            data['city'] = city_name
            merged_data.append(data)

# 将合并后的数据转换为 JSON 格式
merged_json_data = json.dumps(merged_data, indent=4)

# 保存合并后的数据到一个新的 JSON 文件
merged_json_file_path = '/Users/jianing/Documents/Study/Recsys/Hands-On-Recommender-System/recommendations/gpt_data/merged_gpt_travel_data.json'
with open(merged_json_file_path, 'w') as merged_json_file:
    merged_json_file.write(merged_json_data)
```

```python
import pandas as pd
import json

# 读取Excel文件中的所有Sheet
excel_file_path = '/Users/jianing/Documents/Study/Recsys/Hands-On-Recommender-System/data/all_oa_data_2.xlsx'
xls = pd.ExcelFile(excel_file_path)

# 初始化一个空的DataFrame列表
all_sheets_data = []

# 遍历所有Sheet，将其数据添加到DataFrame列表中
for sheet_name in xls.sheet_names:
    sheet_data = pd.read_excel(xls, sheet_name=sheet_name)
    category = sheet_name.replace('df_', '')
    if category.startswith('df_'):
        category = category.replace('df_', '', 1)
    sheet_data['category'] = category
    # 删除名为 'Unnamed: 9', 'Unnamed: 10', 'Unnamed: 11' 的列
    columns_to_remove = ['Unnamed: 9', 'Unnamed: 10', 'Unnamed: 11']
    sheet_data.drop(columns=[col for col in columns_to_remove if col in sheet_data.columns], inplace=True)
    
    all_sheets_data.append(sheet_data)

# 将所有Sheet的数据合并到一个DataFrame中
merged_data = pd.concat(all_sheets_data, ignore_index=True)

# 将合并后的数据转换为JSON格式
json_data = merged_data.to_json(orient='records', indent=4)

# 保存JSON数据到文件
json_file_path = 'merged_oa_data_2.json'
with open(json_file_path, 'w') as json_file:
    json_file.write(json_data)

print(f"JSON data has been saved to {json_file_path}")
```

```python
import json
import pandas as pd

# 读取JSON文件
with open('/Users/jianing/Documents/Study/Recsys/Hands-On-Recommender-System/recommendations/final_data/merged_gpt_travel_data.json', 'r', encoding='utf-8') as file:
    data1 = json.load(file)

with open('/Users/jianing/Documents/Study/Recsys/Hands-On-Recommender-System/recommendations/final_data/merged_oa_data.json', 'r', encoding='utf-8') as file:
    data2 = json.load(file)



# 定义统一的键
keys = [
    "title", "description", "address", "website", "openingHours", "price", 
    "duration", "reachablePublicTransport", "category", "city", 
    "geometry", "ranking", "countryCode"
]

# 将两个文件的数据合并
merged_data = []

def transform_item(item):
    transformed_item = {
        "title": item.get("title", ""),
        "description": item.get("shortText", item.get("description", "")),
        "address": item.get("address", ""),
        "website": item.get("homepage", item.get("website", "")),
        "openingHours": item.get("businessHours", item.get("openingHours", "")),
        "price": item.get("price", ""),
        "duration": item.get("duration", ""),
        "reachablePublicTransport": item.get("reachablePublicTransport", False),
        "category": item.get("category", ""),
        "city": item.get("city", ""),
        "geometry": item.get("geometry", ""),
        "ranking": item.get("ranking", ""),
        "countryCode": item.get("countryCode", "")
    }
    return transformed_item

# 处理第一个文件的数据
for item in data1:
    merged_data.append(transform_item(item))

# 处理第二个文件的数据
for item in data2:
    merged_data.append(transform_item(item))

# 保存合并后的数据
merged_file_path = 'merged_data.json'
with open(merged_file_path, 'w') as file:
    json.dump(merged_data, file, indent=4)

print(f"Merged data has been saved to {merged_file_path}")

```
